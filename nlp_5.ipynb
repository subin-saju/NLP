```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_df=.95,min_df=2,stop_words='english')
dtm=cv.fit_transform(data['Article'])
```

```python
dtm ## cv chose 5477 words from total documents and cv.fit transform words  into numerical data
```

```python
from sklearn.decomposition import LatentDirichletAllocation
lda=LatentDirichletAllocation(n_components=7,random_state=42) ## create a mutinomial pro of 7 topics
lda.fit(dtm)
```

```python
len(cv.get_feature_names_out())
```

```python
# Get the feature names (words)
feature_names = cv.get_feature_names_out()

# Print the feature names
feature_names[54775]
```

```python
len(lda.components_)
```

```python
new=pd.DataFrame(lda.components_)
new ## words and coresponding prob
```

```python
s = lda.components_[0]  ## Topic 0
```

```python
k=s.argsort()[-10:] ## args with highest prob
```

```python
for i in k:
  print(feature_names[i])
## words with high prob in doc 1
```

```python
for  i in range(7) :
    t=lda.components_[i]
    k=t.argsort()[-5:]
    print("######")
    print("#######")
    print("######")
    print(f'topic {i}')
    for j in k:
        print(feature_names[j])
##topics with high prob in each doc
```

```python
from sklearn.decomposition import NMF
nmf_model = NMF(n_components=7,random_state=42)
# Fit the NMF model on the transformed data
nmf_model.fit(cv.fit_transform(data['Article']))
```

```python
nmf_model.components_.shape
```

```python
topic_results = nmf_model.transform(dtm) ## axis 1 = topic , axis 0=  doc
topic_results.shape
new_=pd.DataFrame(topic_results)
new_
```

```python
data['Topic'] = topic_results.argmax(axis=1) ## argmax gives   the index with high prob
```

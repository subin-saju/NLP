```python
import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp(u'Tesla is U.S. company with $6 worth')
for token in doc:
    if not token.is_space:
        print(token.text, token.pos, token.dep)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp(u'spaCy (/speɪˈsiː/ spay-SEE) is an open-source software library for advanced natural language processing, written in the programming languages Python and Cython. The library is published under the MIT license and its main developers are Matthew Honnibal and Ines Montani,')
for token in doc:
    if not token.is_space:
        print(token.text, token.pos_, token.dep_)
```

```python
for sent in doc.sents:
    print(sent)
    print('\n')
    print('\n')
```

```python
for  e in doc.ents:
    print(e.text, e.label_)
```

```python
for noun in doc.noun_chunks:
    print(noun)
```

```python
from spacy import displacy
displacy.render(doc, style='dep', jupyter=True, options={'distance':110})
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
p_stemmer = PorterStemmer()
words = ['run', 'runner', 'running', 'ran', 'runs', 'fairness', 'fairly']
for word in words:
    print(word + ' --> ' + p_stemmer.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
s_stemmer = SnowballStemmer(language='english')
for word in words:
    print(word + ' --> ' + s_stemmer.stem(word))
```

```python
for token in doc:
    print(f'{token.text:{20}}, {token.lemma_:{10}}')
```
